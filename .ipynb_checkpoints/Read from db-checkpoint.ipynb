# Create new table, copy over sorted by date, delete og, rename new table to og name 

```python
import sqlite3
import pandas as pd

# Connect to the database
con = sqlite3.connect(r"C:\Users\Jonat\Documents\MEGAsync\MEGAsync\Github\sp500_data\Scripts\min\min_sp500_market_data.db")

# Define the SQL query to create a new table with sorted data
create_sorted_table_query = """
CREATE TABLE sorted_stock_data AS
SELECT * FROM stock_data
ORDER BY date;
"""

# Execute the query to create the sorted table
con.execute(create_sorted_table_query)

# Commit the transaction
con.commit()

# Drop the original table
drop_original_table_query = """
DROP TABLE IF EXISTS stock_data;
"""

con.execute(drop_original_table_query)

# Rename the sorted table to the original table name
rename_table_query = """
ALTER TABLE sorted_stock_data
RENAME TO stock_data;
"""

con.execute(rename_table_query)

# Commit the transaction
con.commit()

# Close the database connection
con.close()
print("done")
```

```python
# CHANGE date format was NUM > TIMESTAMP

con = sqlite3.connect(r"C:\Users\Jonat\Documents\MEGAsync\MEGAsync\Github\sp500_data\Scripts\min\min_sp500_market_data.db")
# Use ALTER TABLE to change the data type of the 'date' column to TIMESTAMP
con.execute("PRAGMA foreign_keys=off")

# Backup the existing table
con.execute("ALTER TABLE stock_data RENAME TO stock_data_old")

# Recreate the table with the new data type
con.execute("""
    CREATE TABLE stock_data (
        date TIMESTAMP,
        symbol TEXT,
        open REAL,
        high REAL,
        low REAL,
        close REAL,
        adj_close REAL,
        volume INT
    )
""")

# Copy data from the old table to the new table
con.execute("INSERT INTO stock_data SELECT * FROM stock_data_old")

# Drop the old table
con.execute("DROP TABLE stock_data_old")

# Commit the changes
con.commit()
print("done")
```

# 1 min, 1m = max 7 days within last 30 days 

```python
import datetime as dt
from datetime import datetime, timedelta

wk0 = (dt.datetime.today() - timedelta(days=7)).strftime('%Y-%m-%d')
wk1 = (datetime.strptime(wk0, '%Y-%m-%d') - timedelta(days=7)).strftime('%Y-%m-%d')
wk2 = (datetime.strptime(wk1, '%Y-%m-%d') - timedelta(days=14)).strftime('%Y-%m-%d')
wk3 = (datetime.strptime(wk2, '%Y-%m-%d') - timedelta(days=21)).strftime('%Y-%m-%d')

print(f"""
You are running this on: {dt.datetime.today().strftime('%Y-%m-%d')}

Prev 4 time periods:
{wk3} {wk2}
{wk2} {wk1}
{wk1} {wk0}
{wk0} {dt.datetime.today().strftime('%Y-%m-%d')}
""")
```
