```python
import yfinance as yf
def get_stock_data(symbol, start, end):
    data = yf.download(symbol, start=start, end=end)
    data.insert(0, "symbol", symbol)
    
    data.rename(columns={
        "symbol": "symbol",
        "Open": "open",
        "High": "high",
        "Low": "low",
        "Close": "close",
        "Adj Close": "adj_close",
        "Volume": "volume"
    }, inplace=True)

    return data

# Function to format numeric values with commas
def format_with_commas(value):
    if isinstance(value, (int, float)):
        return '{:,.0f}'.format(value)  # Format as integer with commas
    return value  # Keep non-numeric values unchanged

df = get_stock_data('AAPL', '2023-10-01', '2023-10-03')
df['volume']= df['volume'].apply(format_with_commas)
df
```
