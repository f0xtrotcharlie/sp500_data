# Create new table, copy over sorted by date, delete og, rename new table to og name 

```python
import sqlite3
import pandas as pd

# Connect to the database
con = sqlite3.connect(r"C:\Users\Jonat\Documents\MEGAsync\MEGAsync\Github\sp500_data\Scripts\min\min_sp500_market_data.db")

# Define the SQL query to create a new table with sorted data
create_sorted_table_query = """
CREATE TABLE sorted_stock_data AS
SELECT * FROM stock_data
ORDER BY date;
"""

# Execute the query to create the sorted table
con.execute(create_sorted_table_query)

# Commit the transaction
con.commit()

# Drop the original table
drop_original_table_query = """
DROP TABLE IF EXISTS stock_data;
"""

con.execute(drop_original_table_query)

# Rename the sorted table to the original table name
rename_table_query = """
ALTER TABLE sorted_stock_data
RENAME TO stock_data;
"""

con.execute(rename_table_query)

# Commit the transaction
con.commit()

# Close the database connection
con.close()
print("done")
```

# Vacuum

SQLite does not support data compression as a built-in feature. When you write new data to an SQLite database, the file size may not change significantly for several reasons:

Data Pages: SQLite stores data in fixed-size pages (usually 4096 bytes per page by default). When you insert new data, it may not take up a full page, and SQLite may not immediately reclaim the unused space. This can lead to the database file not growing in size even if you are adding new data.

Journaling and Transactions: SQLite uses a journaling mechanism to ensure data consistency and durability. Data is first written to a journal file, and then the changes are committed to the main database file. This process can cause fluctuations in the file size, but it may not immediately reflect the new data until the changes are committed.

Freelist: SQLite maintains a freelist of unused pages. When you delete data, the space is marked as free in the database file, but it's not immediately returned to the operating system or reduced in file size. This free space is available for future insertions.

Database Vacuuming: SQLite provides a "VACUUM" command that can be used to rebuild the database file and reclaim unused space. You can manually run the VACUUM command to reduce the size of the database file. However, this operation can be time-consuming, especially for large databases.

If you want to reduce the size of your SQLite database file while it's still actively being written to, you can periodically run the VACUUM command. Keep in mind that this operation can lock the database and may have some performance impact, so it's important to schedule it during maintenance windows or periods of low activity. Here's how you can run the VACUUM command in SQLite:

```python
import sqlite3

# Open a connection to your database
conn = sqlite3.connect(r"C:\Users\Jonat\Documents\MEGAsync\MEGAsync\Github\sp500_data\Scripts\min\min_sp500_market_data.db")

# Create a cursor
cursor = conn.cursor()

# Run the VACUUM command
cursor.execute('VACUUM')

# Commit the changes and close the connection
conn.commit()
conn.close()
print('Vacuum-ed!')
```
