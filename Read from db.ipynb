```python
#deletes all duplicates
sql_query = """
DELETE FROM stock_data
WHERE (symbol, Date)
IN (SELECT symbol, Date
    FROM stock_data
    GROUP BY symbol, Date
    HAVING COUNT(*) > 1
);
"""

con.execute(sql_query)

pd.read_sql_query("SELECT * from stock_data", con)
```

# Dropping Duplicates from SQL table

```python
import sqlite3
import pandas as pd

#deletes duplicates, keep 1 copy
# connect to the database
con = sqlite3.connect(r"C:\Users\Jonat\Documents\MEGAsync\MEGAsync\Github\sp500_data\sp500_market_data.db")

sql_query = """
DELETE FROM stock_data
WHERE rowid NOT IN (
    SELECT MIN(rowid)
    FROM stock_data
    GROUP BY symbol, date
);
"""

con.execute(sql_query)
con.commit()

# pd.read_sql_query("SELECT * from stock_data", con)
pd.read_sql_query("SELECT * FROM stock_data WHERE Date='2023-10-20T00:00:00'", con)
```
